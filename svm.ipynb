```
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```
n=1000
x1=np.random.normal(5,1,n)
x2=np.random.normal(5,1,n)
female=np.stack((x1,x2),1)
x1=np.random.normal(15,1,n)
x2=np.random.normal(15,1,n)
male=np.stack((x1,x2),1)
fig=plt.figure(figsize=(12,8))
plt.scatter(female[:,0],female[:,1])
plt.scatter(male[:,0],male[:,1])
plt.show()
```

```
data=np.concatenate((male,female))
data.shape
label=np.concatenate((np.array([1]*n),np.array([0]*n)))
label
data.shape
```

```
def SVM():
    model=tf.keras.Sequential([
                               tf.keras.Input(shape=(2)),
                               tf.keras.layers.Dense(1,activation=None)

    ])
    model.compile(loss='hinge',optimizer=tf.keras.optimizers.Adam(0.01),metrics=['accuracy'])
    return model
classifier=SVM()
classifier.fit(data,label,epochs=100)
```

#after training

```
pred=classifier.predict(data)
male=data[np.where(pred>0)[0]]
female=data[np.where(pred<0)[0]]
```

```
print(male)
fig=plt.figure(figsize=(12,8))
plt.scatter(female[:,0],female[:,1])
plt.scatter(male[:,0],male[:,1])
plt.show()
```

```
mm=min(pred[pred>0])
ff=max(pred[pred<0])
fig=plt.figure(figsize=(12,8))
plt.scatter(female[:,0],female[:,1])
plt.scatter(male[:,0],male[:,1])
ind,_=np.where(pred==mm)
sv_male=data[ind]
plt.scatter(sv_male[:,0],sv_male[:,1],c='green')
ind,_=np.where(pred==ff)
sv_female=data[ind]
plt.scatter(sv_female[:,0],sv_female[:,1],c='red')

plt.show()

```
